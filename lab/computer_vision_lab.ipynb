````python

from keras.datasets import mnist
df = mnist.load_data('minst.db')
train,test = df
X_train, y_train = train
X_test, y_test = test

import matplotlib.pyplot as plt
import numpy as np
np.set_printoptions(edgeitems=30, linewidth=100000)
for i in range(5):
    print(y_test[i],'\n') #print label
    print(X_test[i],'\n') # print matrix
    plt.contourf(np.rot90(X_test[i].transpose())) #make contoru plot of the matrix value
    plt.show()
````

````python
#2
print(X_train.shape)
print(X_test.shape)
print(X_train[0].shape)
print(X_test[0].shape)
print(y_train.shape)
print(y_test.shape)
````

The training set X_train has 60,000 and the test set X_test has 10,000 images. Each image like x_train[i] or X_test[i] is a 28x28 matrix. The labels y_train and y_test are arrays of size 60,000 and 10,000, where each value is the actual digit (0-9) shown in the image

````python
#3
import pandas as pd
import numpy as np
import os
reload = True

if reload: # as long as reload is set to True
  Z_train = X_train.reshape((X_train.shape[0], 784))
  Z_test = X_test.reshape((X_test.shape[0], 784))
  # Convert to DataFrames for saving
  df_Z_train = pd.DataFrame(Z_train)
  df_Z_test = pd.DataFrame(Z_test)
  # Save Z-train csv
  df_Z_train.to_csv("Z_train.csv", index=False)
  df_Z_test.to_csv("Z_test.csv", index=False)
else:
    df_Z_train = pd.read_csv("Z_train.csv") #loads csv
    df_Z_test = pd.read_csv("Z_test.csv")
    # Convert back to numpy arrays if needed
    Z_train = df_Z_train.to_numpy()
    Z_test = df_Z_test.to_numpy()
````

````python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import numpy as np

# used smaller data for speed so execution wouldnt take a long time
Z_train_small = Z_train[:10000]
y_train_small = y_train[:10000]
Z_test_small = Z_test[:1000]
y_test_small = y_test[:1000]

# starting by trying k from 1 to 50
k_grid = np.arange(1, 51)
accuracy_scores = {}

for k in k_grid:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(Z_train_small, y_train_small)
    y_pred = knn.predict(Z_test_small)
    acc = accuracy_score(y_test_small, y_pred)
    accuracy_scores[k] = acc
    print(f"k = {k}, accuracy = {acc:.4f}")

# finds the best k
best_k = max(accuracy_scores, key=accuracy_scores.get)
print("\n Best k:", best_k)
print("Best accuracy:", accuracy_scores[best_k])

# plots the result
plt.figure(figsize=(10, 6))
plt.plot(k_grid, [accuracy_scores[k] for k in k_grid], marker='o', linestyle='-')
plt.xlabel('k (number of neighbors)')
plt.ylabel('Accuracy on test set')
plt.title(f'Accuracy vs. k (Best k = {best_k})')
plt.grid(True)
plt.show()
````

````python
#5
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, accuracy_score


knn_best = KNeighborsClassifier(n_neighbors=best_k)
knn_best.fit(Z_train, y_train)
y_pred = knn_best.predict(Z_test)
acc = accuracy_score(y_test, y_pred)
print(f"Best k = {best_k}, accuracy = {acc:.4f}")

#confusion matrix:
cm = confusion_matrix(y_test, y_pred)
display = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=np.arange(10))
display.plot(cmap='Blues',xticks_rotation='vertical')
plt.title("Confusion Matrix (k= {best_k})")
plt.show()
````

7. A color photo can be stored as three separate 2D matrices: one for Red, one for Green, and one for Blue. Each matrix has pixel values from 0 to 255 that shows us how strong that color is at each point. So if the image is 100×100 pixels, we end up with a 100×100×3 array. To turn this into a tabular data we can flatten the image to a one long row. For a 100×100 image it will give us 100×100×3 = 30,000 values which results in each image becomes a 1×30,000 vector. Then we can feed it into machine learning models like any regular dataset.
